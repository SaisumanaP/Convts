```python
import torch
```

```python
from criterion.combination import CombinationLoss
from criterion.sdr import NegWeightedSDR
```

```python
batch_size = 16
n_sources = 4
in_channels, T = 2, 32
```

```python
criterion = NegWeightedSDR()
combination_criterion = CombinationLoss(criterion, min_pair=1, max_pair=n_sources-1)
```

```python
input = torch.randn(batch_size, n_sources, in_channels, T)
target = torch.randn(batch_size, n_sources, in_channels, T)

loss = combination_criterion(input, target, reduction=False, batch_mean=False)
print(loss.size()) # (batch_size, n_combinations), where n_combinations = 4C1+4C2+4C3 = 14
```

```python
input = torch.randn(batch_size, n_sources, in_channels, T)
target = input

loss = combination_criterion(input, target, batch_mean=True)
print(loss)
```
