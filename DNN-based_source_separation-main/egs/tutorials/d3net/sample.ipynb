```python
import torch
```

```python
from models.d3net import D3Net
```

```python
torch.manual_seed(111)
```

```python
batch_size, in_channels, n_bins, n_frames = 4, 2, 2049, 256

input = torch.randn(batch_size, in_channels, n_bins, n_frames) # random input
input = torch.abs(input)
```

```python
model_path = "./model.pth"
config_path = "./config/paper/vocals.yaml" # "./config/paper/vocals.yaml", "./config/paper/drums.yaml", "./config/paper/bass.yaml", "./config/paper/other.yaml"
```

```python
model = D3Net.build_from_config(config_path)
```

```python
print(model)
print("# Parameters: {}".format(model.num_parameters))
```

```python
if torch.cuda.is_available():
    input = input.cuda()
    model.cuda()
    print("Use CUDA.")
else:
    print("Doesn't use CUDA.")
```

```python
with torch.no_grad():
    output = model(input)
```

```python
print(input.size(), output.size())
```

```python
config = model.get_package()
```

```python
config['state_dict'] = model.state_dict()
torch.save(config, model_path)
```

```python
saved_model = D3Net.build_model(model_path)
config = torch.load(model_path, map_location=lambda storage, loc: storage)
saved_model.load_state_dict(config['state_dict'])
```
