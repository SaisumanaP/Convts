```python
import IPython.display as ipd
```

```python
import torch
import torchaudio
from torchaudio.transforms import GriffinLim, Spectrogram
```

## Single-channel

```python
y, sr = torchaudio.load("../../../dataset/sample-song/single-channel/sample-1_electric-guitar_8000.wav")
```

```python
print(y.size(), sr)
```

```python
T = y.size(1)
```

```python
ipd.Audio(y, rate=sr)
```

```python
fft_size, hop_size = 2048, 1024
```

```python
spectrogram = Spectrogram(
    n_fft=fft_size,
    hop_length=hop_size,
    power=2
)
griffin_lim = GriffinLim(
    n_fft=fft_size,
    hop_length=hop_size,
    power=2
)
```

```python
if torch.cuda.is_available():
    spectrogram = spectrogram.cuda()
    griffin_lim = griffin_lim.cuda()
    y = y.cuda()
```

```python
Y = spectrogram(y)
y_hat = griffin_lim(Y) # returns waveform
```

```python
y_hat.size()
```

```python
y_hat = y_hat[..., :T].cpu()
ipd.Audio(y_hat, rate=sr)
```
